# TROUBLESHOOTING_CONTENT

## Trouble Shooting
The following section provides general information about troubleshooting courseware produced by Databricks Academy. Each section correlates to a specific error message that an end-user may receive with direct links to more advanced content.

### Admin Required

Databricks Academy often requires the student to have admin privileges in a workspace to effect changes. In some cases, this is a byproduct of working through the courseware. However, in many cases, this is to configure the environment for users that would otherwise not require admin privileges.

For more current information, please <a href="https://files.training.databricks.com/static/troubleshooting.html#admin-required" target="_blank">Troubleshooting Admin Requirements</a>

### Workspace Not Configured

Some of the content produced by Databricks Academy requires configuration (or re-configuration) of the workspace and its assets. This configuration includes complete tasks like creating cluster policies, cluster pools, granting specific user permissions, and other such tasks.

This error generally indicates that the notebook Workspace-Setup was not executed.

For more current information, please <a href="https://docs.databricks.com/aws/en/workspace/" target="_blank">Troubleshooting Workspace Setup</a>

### Spark Version

To ensure that all labs execute as expected, Databricks Academy requires that the content provided be executed against a cluster configured with a specific runtime. Typically this is an LTS (long-term supported) version.

This course is tested against and requires one of the following DBRs (Databricks Runtime): **`17.3.x-cpu-ml-scala2.13`**

For more current information, please <a href="https://spark.apache.org/docs/latest/index.html" target="_blank">Troubleshooting the Spark Version</a>

### Cannot write to DBFS

DBFS stands for Databricks File System and is a virtual file system that Databricks Academy uses to read data from and write data to. There are two primary locations, and depending on the courseware, some tertiary locations.

The first location is **dbfs:/mnt/dbacademy-datasets/**, which is a storage location for datasets used by this course. Datasets are downloaded from our data repositories and copied to this location to provide faster, in-region access to the course's datasets. Files written to this location are expected to be treated as read-only and shared by all consumers of this courseware.

The second location is **dbfs:/mnt/dbacademy-users/**, a storage location for user-specific datasets and misc files generated by working through the courseware. This folder is subdivided by user and then by course, creating a per-user-per-course working directory.

In some cases, tertiary data may be written to **dbfs:/user/** as a default storage location for Spark and Spark tables. Like dbacademy-users, this folder is subdivided by user.

For more current information, please <a href="https://docs.databricks.com/aws/en/dbfs" target="_blank">Troubleshooting DBFS Writes</a>

### Cannot Install Libraries

Databricks Academy provides custom libraries to facilitate the development of its courseware. Your cluster requires access to these remotely hosted libraries so that they can be installed into your cluster.

Various companies restrict access to external sites to limit data exfiltration and address other security concerns, such as leaking potentially sensitive or proprietary information.

For more current information, please see <a href="https://docs.databricks.com/aws/en/libraries/" target="_blank">Troubleshooting Library Installation</a>

### Requires Unity Catalog

Unity Catalog (UC) is a service provided by Databricks. This courseware requires that the workspace be UC Enabled. However, not all workspaces have (or can have) this feature enabled.

For more current information, please see <a href="https://learn.microsoft.com/en-us/azure/databricks/data-governance/unity-catalog/get-started" target="_blank">Troubleshooting Unity Catalog Requirements</a>

### **Issue: Loading the Widget Takes Longer than Expected**
#### **Possible Cause**:
- The notebook is running on an incompatible Databricks Runtime version.
- The widget is too large, leading to longer loading times.
- The internet connection is slow, causing delays in loading.
- A transient issue in the current notebook session or cached state.
#### **Suggested Solution**:
1. **Verify Databricks Runtime Version**:
   - Make sure your cluster is running on **DBR 17.3 ML**
2. **Re-run the Cell**:
   - Sometimes, re-running the cell can resolve issues related to temporary glitches.
3. **Wait for the Widget to Load**:
   - If the widget is large or your internet connection is slow, allow additional time for the widget to load.
4. **Install Required Dependencies**:
   - Ensure that you have the required dependencies installed. Run the following command to install `pygwalker`:
   %pip install pygwalker==0.1.11
5. **Clone and Rerun the Notebook**:
   - If none of the above steps resolve the issue, **clone this notebook** and rerun the cloned version. This action often helps clear any cached states or transient issues in the current notebook session.
6. **Contact Databricks Support**:
   - If the problem persists, reach out to Databricks support for further assistance.
